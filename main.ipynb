## Cell 4: Posterior Predictive Check - Model vs Real Data

```python
# Compare model predictions vs real data

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyddm import Model, Sample, Fittable
from pyddm.models import DriftConstant, NoiseConstant, BoundConstant, OverlayNonDecision, ICPoint
from pyddm.functions import fit_adjust_model
from pyddm.models.loss import LossRobustLikelihood

# Use the custom drift class from modeling module
DriftDistractor = ddm.DriftDistractor

# Create sample from data
sample = Sample.from_pandas_dataframe(df_modeling, rt_column_name="rt", choice_column_name="correct")

# Fit the winning model (Drift-varying)
print("Fitting Drift-varying model to full dataset...")
model = Model(
    name="Drift-varying",
    drift=DriftDistractor(v_dutch=Fittable(minval=-5, maxval=5), v_english=Fittable(minval=-5, maxval=5)),
    bound=BoundConstant(B=Fittable(minval=0.5, maxval=3.0)),
    IC=ICPoint(x0=0.0),
    overlay=OverlayNonDecision(nondectime=Fittable(minval=0.1, maxval=0.5)),
    noise=NoiseConstant(noise=1),
    dx=0.005, dt=0.01, T_dur=3
)

fitted_model = fit_adjust_model(sample=sample, model=model, lossfunction=LossRobustLikelihood, verbose=False)

# Extract fitted parameters
params = fitted_model.parameters()
v_dutch = float(params['drift']['v_dutch'])
v_english = float(params['drift']['v_english'])
B = float(params['bound']['B'])
t0 = float(params['overlay']['nondectime'])

print(f"\nFitted Parameters:")
print(f"  v_dutch: {v_dutch:.4f}")
print(f"  v_english: {v_english:.4f}")
print(f"  B: {B:.4f}")
print(f"  t0: {t0:.4f}")
```

```python
# Generate simulated data from fitted model

print("Generating model predictions...")

# Solve model for each condition
sol_dutch = fitted_model.solve(conditions={"distractor_language": "dutch"})
sol_english = fitted_model.solve(conditions={"distractor_language": "english"})

# Resample to get simulated trials
n_dutch = len(df_modeling[df_modeling["distractor_language"] == "dutch"])
n_english = len(df_modeling[df_modeling["distractor_language"] == "english"])

sim_dutch = sol_dutch.resample(n_dutch)
sim_english = sol_english.resample(n_english)

print(f"Simulated {n_dutch} Dutch trials and {n_english} English trials")
```

```python
# Plot 1: RT Distributions - Model vs Data

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Dutch condition
ax1 = axes[0]
emp_dutch = df_modeling[(df_modeling["distractor_language"] == "dutch") & (df_modeling["correct"] == 1)]["rt"]
sns.kdeplot(emp_dutch, ax=ax1, label="Empirical", color="blue", linewidth=2)
sns.kdeplot(sim_dutch.corr, ax=ax1, label="Model", color="blue", linestyle="--", linewidth=2)
ax1.set_xlabel("Reaction Time (s)")
ax1.set_ylabel("Density")
ax1.set_title("Dutch Distractor - Correct Trials")
ax1.legend()
ax1.grid(True, alpha=0.3)

# English condition
ax2 = axes[1]
emp_english = df_modeling[(df_modeling["distractor_language"] == "english") & (df_modeling["correct"] == 1)]["rt"]
sns.kdeplot(emp_english, ax=ax2, label="Empirical", color="orange", linewidth=2)
sns.kdeplot(sim_english.corr, ax=ax2, label="Model", color="orange", linestyle="--", linewidth=2)
ax2.set_xlabel("Reaction Time (s)")
ax2.set_ylabel("Density")
ax2.set_title("English Distractor - Correct Trials")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.suptitle("Posterior Predictive Check: Model vs Empirical RT Distributions", fontsize=14)
plt.tight_layout()
plt.savefig("generated_figures/figure7_posterior_predictive_rt.png", dpi=150, bbox_inches='tight')
print("Saved: generated_figures/figure7_posterior_predictive_rt.png")
plt.show()
```

```python
# Plot 2: Cumulative Distribution Functions (CDF)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Dutch CDF
ax1 = axes[0]
emp_dutch_sorted = np.sort(emp_dutch)
emp_dutch_cdf = np.arange(1, len(emp_dutch_sorted) + 1) / len(emp_dutch_sorted)
sim_dutch_sorted = np.sort(sim_dutch.corr)
sim_dutch_cdf = np.arange(1, len(sim_dutch_sorted) + 1) / len(sim_dutch_sorted)

ax1.plot(emp_dutch_sorted, emp_dutch_cdf, label="Empirical", color="blue", linewidth=2)
ax1.plot(sim_dutch_sorted, sim_dutch_cdf, label="Model", color="blue", linestyle="--", linewidth=2)
ax1.set_xlabel("Reaction Time (s)")
ax1.set_ylabel("Cumulative Probability")
ax1.set_title("Dutch Distractor - CDF")
ax1.legend()
ax1.grid(True, alpha=0.3)

# English CDF
ax2 = axes[1]
emp_english_sorted = np.sort(emp_english)
emp_english_cdf = np.arange(1, len(emp_english_sorted) + 1) / len(emp_english_sorted)
sim_english_sorted = np.sort(sim_english.corr)
sim_english_cdf = np.arange(1, len(sim_english_sorted) + 1) / len(sim_english_sorted)

ax2.plot(emp_english_sorted, emp_english_cdf, label="Empirical", color="orange", linewidth=2)
ax2.plot(sim_english_sorted, sim_english_cdf, label="Model", color="orange", linestyle="--", linewidth=2)
ax2.set_xlabel("Reaction Time (s)")
ax2.set_ylabel("Cumulative Probability")
ax2.set_title("English Distractor - CDF")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.suptitle("Posterior Predictive Check: Cumulative Distribution Functions", fontsize=14)
plt.tight_layout()
plt.savefig("generated_figures/figure8_posterior_predictive_cdf.png", dpi=150, bbox_inches='tight')
print("Saved: generated_figures/figure8_posterior_predictive_cdf.png")
plt.show()
```

```python
# Plot 3: Quantile-Quantile Plot

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

quantiles = [0.1, 0.3, 0.5, 0.7, 0.9]

# Dutch Q-Q
ax1 = axes[0]
emp_q_dutch = np.quantile(emp_dutch, quantiles)
sim_q_dutch = np.quantile(sim_dutch.corr, quantiles)
ax1.scatter(emp_q_dutch, sim_q_dutch, s=100, c='blue', zorder=3)
min_val = min(emp_q_dutch.min(), sim_q_dutch.min()) - 0.05
max_val = max(emp_q_dutch.max(), sim_q_dutch.max()) + 0.05
ax1.plot([min_val, max_val], [min_val, max_val], 'r--', label='Perfect fit')
for i, q in enumerate(quantiles):
    ax1.annotate(f'{int(q*100)}%', (emp_q_dutch[i], sim_q_dutch[i]), textcoords="offset points", xytext=(5, 5))
ax1.set_xlabel("Empirical Quantiles (s)")
ax1.set_ylabel("Model Quantiles (s)")
ax1.set_title("Dutch Distractor - Q-Q Plot")
ax1.legend()
ax1.grid(True, alpha=0.3)

# English Q-Q
ax2 = axes[1]
emp_q_english = np.quantile(emp_english, quantiles)
sim_q_english = np.quantile(sim_english.corr, quantiles)
ax2.scatter(emp_q_english, sim_q_english, s=100, c='orange', zorder=3)
min_val = min(emp_q_english.min(), sim_q_english.min()) - 0.05
max_val = max(emp_q_english.max(), sim_q_english.max()) + 0.05
ax2.plot([min_val, max_val], [min_val, max_val], 'r--', label='Perfect fit')
for i, q in enumerate(quantiles):
    ax2.annotate(f'{int(q*100)}%', (emp_q_english[i], sim_q_english[i]), textcoords="offset points", xytext=(5, 5))
ax2.set_xlabel("Empirical Quantiles (s)")
ax2.set_ylabel("Model Quantiles (s)")
ax2.set_title("English Distractor - Q-Q Plot")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.suptitle("Quantile-Quantile Comparison: Model vs Data", fontsize=14)
plt.tight_layout()
plt.savefig("generated_figures/figure9_qq_plot.png", dpi=150, bbox_inches='tight')
print("Saved: generated_figures/figure9_qq_plot.png")
plt.show()
```

```python
# Plot 4: Accuracy Comparison

fig, ax = plt.subplots(figsize=(8, 5))

# Empirical accuracy
emp_acc_dutch = df_modeling[df_modeling["distractor_language"] == "dutch"]["correct"].mean()
emp_acc_english = df_modeling[df_modeling["distractor_language"] == "english"]["correct"].mean()

# Model predicted accuracy
sim_acc_dutch = len(sim_dutch.corr) / (len(sim_dutch.corr) + len(sim_dutch.err))
sim_acc_english = len(sim_english.corr) / (len(sim_english.corr) + len(sim_english.err))

x = np.arange(2)
width = 0.35

bars1 = ax.bar(x - width/2, [emp_acc_dutch, emp_acc_english], width, label='Empirical', color=['blue', 'orange'], alpha=0.7)
bars2 = ax.bar(x + width/2, [sim_acc_dutch, sim_acc_english], width, label='Model', color=['blue', 'orange'], hatch='//', alpha=0.5)

ax.set_ylabel('Accuracy')
ax.set_title('Accuracy Comparison: Model vs Empirical')
ax.set_xticks(x)
ax.set_xticklabels(['Dutch', 'English'])
ax.legend()
ax.set_ylim(0.8, 1.0)
ax.grid(True, axis='y', alpha=0.3)

# Add value labels
for bar, val in zip(bars1, [emp_acc_dutch, emp_acc_english]):
    ax.annotate(f'{val:.3f}', (bar.get_x() + bar.get_width()/2, bar.get_height()), 
                ha='center', va='bottom', fontsize=10)
for bar, val in zip(bars2, [sim_acc_dutch, sim_acc_english]):
    ax.annotate(f'{val:.3f}', (bar.get_x() + bar.get_width()/2, bar.get_height()), 
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig("generated_figures/figure10_accuracy_comparison.png", dpi=150, bbox_inches='tight')
print("Saved: generated_figures/figure10_accuracy_comparison.png")
plt.show()

print(f"\nAccuracy Summary:")
print(f"  Dutch  - Empirical: {emp_acc_dutch:.4f}, Model: {sim_acc_dutch:.4f}")
print(f"  English - Empirical: {emp_acc_english:.4f}, Model: {sim_acc_english:.4f}")
```

```python
# Plot 5: Summary Statistics Comparison

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

conditions = ['Dutch', 'English']
x = np.arange(len(conditions))
width = 0.35

# Mean RT
ax1 = axes[0]
emp_means = [emp_dutch.mean(), emp_english.mean()]
sim_means = [np.mean(sim_dutch.corr), np.mean(sim_english.corr)]
ax1.bar(x - width/2, emp_means, width, label='Empirical', color='steelblue')
ax1.bar(x + width/2, sim_means, width, label='Model', color='coral')
ax1.set_ylabel('Mean RT (s)')
ax1.set_title('Mean RT')
ax1.set_xticks(x)
ax1.set_xticklabels(conditions)
ax1.legend()
ax1.grid(True, axis='y', alpha=0.3)

# SD RT
ax2 = axes[1]
emp_sds = [emp_dutch.std(), emp_english.std()]
sim_sds = [np.std(sim_dutch.corr), np.std(sim_english.corr)]
ax2.bar(x - width/2, emp_sds, width, label='Empirical', color='steelblue')
ax2.bar(x + width/2, sim_sds, width, label='Model', color='coral')
ax2.set_ylabel('SD RT (s)')
ax2.set_title('RT Standard Deviation')
ax2.set_xticks(x)
ax2.set_xticklabels(conditions)
ax2.legend()
ax2.grid(True, axis='y', alpha=0.3)

# Median RT
ax3 = axes[2]
emp_medians = [np.median(emp_dutch), np.median(emp_english)]
sim_medians = [np.median(sim_dutch.corr), np.median(sim_english.corr)]
ax3.bar(x - width/2, emp_medians, width, label='Empirical', color='steelblue')
ax3.bar(x + width/2, sim_medians, width, label='Model', color='coral')
ax3.set_ylabel('Median RT (s)')
ax3.set_title('Median RT')
ax3.set_xticks(x)
ax3.set_xticklabels(conditions)
ax3.legend()
ax3.grid(True, axis='y', alpha=0.3)

plt.suptitle('Summary Statistics: Model vs Empirical Data', fontsize=14)
plt.tight_layout()
plt.savefig("generated_figures/figure11_summary_stats.png", dpi=150, bbox_inches='tight')
print("Saved: generated_figures/figure11_summary_stats.png")
plt.show()

# Print summary table
print("\n" + "="*60)
print("SUMMARY STATISTICS COMPARISON")
print("="*60)
print(f"{'Metric':<15} {'Condition':<10} {'Empirical':<12} {'Model':<12} {'Diff':<10}")
print("-"*60)
for i, cond in enumerate(conditions):
    print(f"{'Mean RT':<15} {cond:<10} {emp_means[i]:<12.4f} {sim_means[i]:<12.4f} {abs(emp_means[i]-sim_means[i]):<10.4f}")
    print(f"{'SD RT':<15} {cond:<10} {emp_sds[i]:<12.4f} {sim_sds[i]:<12.4f} {abs(emp_sds[i]-sim_sds[i]):<10.4f}")
    print(f"{'Median RT':<15} {cond:<10} {emp_medians[i]:<12.4f} {sim_medians[i]:<12.4f} {abs(emp_medians[i]-sim_medians[i]):<10.4f}")
    print("-"*60)
```
